```python
#tk = th.Thicket.from_caliperreader(glob("cali_files/*Random.cali"))
files = glob("cali_files/*Random.cali")
print(files)

exclude_files = ['cali_files/mergeSortMPI-p512-a67108864-Random.cali',
                 'cali_files/mergeSortMPI-p512-a65536-Random.cali']

# Filter out excluded files
filtered_files = [file for file in files if file not in exclude_files]

tk = th.Thicket.from_caliperreader((filtered_files))
```

```python
tk.metadata
```

```python
gb = tk.groupby("InputSize")
```

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis="columns",
    metadata_key="num_procs"
)
```

```python
ctk.dataframe
```

```python
# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop(["nid", "spot.channel", "Min time/rank", "Max time/rank", "Avg time/rank", "Variance time/rank"], level=1, axis=1)
ctk.dataframe = ctk.dataframe.drop(("node", ""), axis=1)

# Rename index
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
```

```python
ctk.dataframe
```

```python
# Define common variables
function_names = set(ctk.dataframe.index.get_level_values("name"))
threads = sorted(list(set(ctk.dataframe.index.get_level_values("num_procs"))))
matrix_sizes = ctk.dataframe.columns.tolist()
```

```python
def plot_mpl(df, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(matrix_sizes)
    plt.grid(True)
    #plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
ctk.dataframe.loc["main"]
```

```python
for name in function_names:
    plot_mpl(
        df=ctk.dataframe.loc[name],
        title=name,
        xlabel="Threads",
        ylabel="Time (seconds)",
    )
    #
    #
    #
    #
```
